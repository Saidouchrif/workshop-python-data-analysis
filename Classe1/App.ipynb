## Pourquoi NumPy ?

NumPy est une bibliothèque fondamentale pour le calcul scientifique en Python.

Avantages :
- Très rapide (écrit en C)
- Support des tableaux multidimensionnels (ndarray)
- Opérations mathématiques optimisées
- Base de la majorité des bibliothèques Data Science (Pandas, Scikit-learn, etc.)

## 1. Définir un vecteur (Array 1D)

On peut créer un tableau NumPy à partir d'une liste Python.

```python
import numpy as np
A=np.array([1,2,3])
print(A)
print(type(A))
```

## 2. Définir une matrice (Array 2D)

NumPy permet de créer des matrices multidimensionnelles.

```python
A = np.array([1, 2, 3])
B = np.zeros((2, 3))
C = np.ones((2, 3))

print("A:\n", A, "\n")
print("B:\n", B, "\n")
print("C:\n", C)
```

## 3. Remplir une matrice avec une valeur donnée

La fonction np.full() permet de remplir un tableau avec une valeur spécifique.

```python
D = np.full((2, 2), 659)
print(D)

E = np.full_like(B, 4)
print(E)
```

## 4. Génération de tableaux aléatoires

NumPy permet de générer des valeurs aléatoires selon différentes distributions.

```python
A = np.random.randn(2, 3)   # distribution normale
B = np.random.rand(2, 3)    # distribution uniforme
C = np.random.randint(0, 10, (2, 3))  # entiers aléatoires

print("Randn:\n", A, "\n")
print("Rand:\n", B, "\n")
print("RandInt:\n", C)
```

## 5. Informations sur un tableau

On peut récupérer :
- shape : dimensions
- ndim : nombre de dimensions
- size : nombre total d’éléments

```python
A = np.array([[1, 2, 3], [4, 5, 6]])

print("Shape:", A.shape)
print("Dimensions:", A.ndim)
print("Size:", A.size)
```
